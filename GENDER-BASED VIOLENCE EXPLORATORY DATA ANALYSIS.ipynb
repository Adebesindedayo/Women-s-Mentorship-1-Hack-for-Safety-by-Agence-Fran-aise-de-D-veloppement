Missing Values
No missing value in the train and test dataset

```python
## Here we will check the percentage of nan values present in each feature of train
## 1 -step make the list of features which has missing values
features_with_na=[features for features in train.columns if train[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(train[feature].isnull().mean(), 4),  ' % missing values')
```

```python
## Here we will check the percentage of nan values present in each feature of test
## 1 -step make the list of features which has missing values
features_with_na=[features for features in test.columns if test[features].isnull().sum()>1]
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(test[feature].isnull().mean(), 4),  ' % missing values')
```

```python
sns.set(rc= {"figure.figsize": (12, 15)})
ax = sns.countplot(y='Education',data=GBV,)# using a countplot


total = len(GBV['Education'])
for p in ax.patches: # Using a for loop to determine the percentage
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
plt.yticks(size=20)
plt.xticks(size=20)

plt.title("Percentage of the Education that experience gender-based violence",fontsize=20)
plt.savefig('Percentage of Education.png')
```
