**PERCENTAGE OF EACH CATEGORICAL FEATURES THAT WAS VIOLATED**

```python
GBV=train.loc[(train.target ==1)]
```

```python
sns.set(rc= {"figure.figsize": (10, 10)})
ax = sns.countplot(y='Education',data=GBV,)# using a countplot
total = len(GBV['Education'])
for p in ax.patches: # Using a for loop to determine the percentage
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
plt.yticks(size=20)
plt.xticks(size=20)

plt.title("Percentage of the Education that experience gender-based violence",fontsize=20)
plt.savefig('Percentage of Education.png')
```
